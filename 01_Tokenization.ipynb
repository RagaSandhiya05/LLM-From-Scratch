**TOKENIZATION**

**Input Text**

```python
text = """
Artificial Intelligence is changing the world.
Large Language Models understand human language.
ChatGPT is an example of an LLM.
"""
print("Original Text:\n")
print(text)
```

**Convert to Lowercase**

```python
text = text.lower()
print("Lowercase Text:\n")
print(text)
```

**Remove Punctuation**

```python
import string
clean_text = text.translate(str.maketrans('', '', string.punctuation))
print("Clean Text:\n")
print(clean_text)
```

**Word Tokenization**

```python
word_tokens = clean_text.split()
print("Word Tokens:\n")
print(word_tokens)
```

**Count Tokens**

```python
print("Total Tokens : " , len(word_tokens))
```

**Display Each Token**

```python
print("Tokens with Position:\n")
for index , token in enumerate(word_tokens) :
  print(f"Token {index + 1} : {token}")
```

**Character Tokenization**

```python
characters = list(clean_text)
print(characters)
```

**Compare Word vs Character Tokens**

```python
print("Number of Word Tokens : " , len(word_tokens))
print("Number of Character Tokens : " , len(characters))
```

**Simple Next-Word Pairs**

```python
print("Next Word Training Piars:\n")
for i in range(len(word_tokens) - 1) :
  print(word_tokens[i] , "--->" , word_tokens[i + 1])
```

**First and Last Token**

```python
print("First Token : " , word_tokens[0])
print("Last Token : " , word_tokens[-1])
```

**Inroduction to Byte Pair Encoding (BPE)**

**Instead of treating a whole word as one token, BPE splits words into smaller pieces (subwords).**

**Example 1**

```python
word = "playing"
print("Original Word : ")
print(word)
bpe_tokens = ["play" , "ing"]
print("\nBPE Tokens : ")
print(bpe_tokens)
```

**Example 2**

```python
word = "unhappiness"
print("Original Word : ")
print(word)
bpe_tokens = ["un" , "happi" , "ness"]
print("\nBPE Tokens : ")
print(bpe_tokens)
```
