**POSITIONAL ENCODING**

**Import Libraries**

```python
import torch
import math
torch.manual_seed(42)
print("PyTorch Version : " , torch.__version__)
```

**Sample Tokens and Embeddings**

```python
tokens = ["I" , "love" , "Python"]
embeddings = torch.tensor([
    [1.0 , 0.5 , 0.2 , 0.1] ,
    [0.8 , 1.0 , 0.4 , 0.3] ,
    [0.2 , 0.1 , 1.0 , 0.9]
])
print("Tokens : ")
print(tokens)

print("\nEmbeddings : ")
print(embeddings)
```

**Positional Encoding Function**

```python
def positional_encoding(sequence_length , embedding_dim) :
  pe = torch.zeros(sequence_length , embedding_dim)
  for pos in range(sequence_length) :
    for i in range(0 , embedding_dim , 2) :
      pe[pos , i] = math.sin(pos / (10000 ** (i / embedding_dim)))
      if i + 1 < embedding_dim :
        pe[pos , i + 1] = math.cos(pos / (10000 ** (i / embedding_dim)))
  return pe
```

**Generate Positional Encoding**

```python
sequence_length = len(tokens)
embedding_dim = embeddings.shape[1]
pe = positional_encoding(sequence_length , embedding_dim)
print("Positional Encoding : \n")
print(pe)
```

**Add Positional Encoding to Embeddings**

```python
encoded_embeddings = embeddings + pe
print("Embeddings After Positional Encoding : \n")
print(encoded_embeddings)
```

**Compare Before and After**

```python
for token , original , encoded in zip(tokens , embeddings , encoded_embeddings) :
  print("=" * 40)
  print("Token : " , token)
  print("\nOriginal Embedding : ")
  print(original)
  print("\nAfter Positional Encoding : ")
  print(encoded)
```

**Token Positions**

```python
print("Token Positions\n")
for position , token in enumerate(tokens) :
  print(f"Position {position} -> {token}")
```

**Why Positional Encoding ?**

```python
sentence1 = ["Dog" , "bites" , "man"]
sentence2 = ["Man" , "bites" , "dog"]
print("Sentence 1 : " , sentence1)
print("Sentence 2 : " , sentence2)
print("\nBoth contain the same wordss.")
print("Positional Encoding helps the model to understand the different word order.")
```

**Display Positional Encoding Matrix**

```python
print("Positional Encoding Matrix\n")
for position , vector in  enumerate(pe) :
  print(f"Position {position} :")
  print(vector)
  print()
```
