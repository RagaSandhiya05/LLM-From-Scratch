**MULTI-HEAD ATTENTION**

**Import Libraries**

```python
import torch
import torch.nn.functional as F
torch.manual_seed(42)
print("PyTorch Version : " , torch.__version__)
```

**Sample Embeddings**

```python
tokens = ["I" , "love" , "Python"]
embeddings = torch.tensor([
    [1.0 , 0.5 , 0.2 , 0.1] ,
    [0.8 , 1.0 , 0.4 , 0.3] ,
    [0.2 , 0.1 , 1.0 , 0.9]
])
print("Tokens : ")
print(tokens)

print("'\nEmbeddings : ")
print(embeddings)
```

**Self-Attention Function**

```python
def self_attention(embeddings) :
  embedding_dim = embeddings.shape[1]
  Wq = torch.randn(embedding_dim , embedding_dim)
  Wk = torch.randn(embedding_dim , embedding_dim)
  Wv = torch.randn(embedding_dim , embedding_dim)

  Q = embeddings @ Wq
  K = embeddings @ Wk
  V = embeddings @ Wv

  scores = Q @ K.T

  scores = scores / (embedding_dim ** 0.5)

  attention = F.softmax(scores , dim = -1)

  context = attention @ V
  return context
```

**Head 1**

```python
head1 = self_attention(embeddings)
print("Head 1 Output :\n")
print(head1)
```

**Head 2**

```python
head2 = self_attention(embeddings)
print("Head 2 Output :\n")
print(head2)
```

**Concatenate Heads**

```python
multi_head_output = torch.cat((head1 , head2) , dim = 1)
print("Concatenated Output :\n")
print(multi_head_output)
```

**Check Shape**

```python
print("Head 1 Shape : " , head1.shape)
print("Head 2 Shape : " , head2.shape)
print("Combined Shape : " , multi_head_output.shape)
```

**Output Projection**

```python
projection = torch.randn(8 , 4)
final_output = multi_head_output @ projection
print("Projected Output :\n")
print(final_output)
```

**Display Final Output**

```python
for token , vector in zip(tokens , final_output) :
  print("=" * 40)
  print("Token : " , token)
  print("Output Vector : ")
  print(vector)
```

**Compare Single vs Multi Head**

```python
print("Single Head Shape : " , head1.shape)

print("Multi Head Shape : " , multi_head_output.shape)

print("Projected Shape : " , final_output.shape)
```
