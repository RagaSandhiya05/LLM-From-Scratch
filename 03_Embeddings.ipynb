**WORD EMBEDDINGS**

**Import PyTorch**

```python
import torch
import torch.nn as nn
print("PyTorch Version : " , torch.__version__)
```

**Create Sample Vocabulary**

```python
words = [
    "artificial" ,
    "intelligence" ,
    "is" ,
    "changing" ,
    "the" ,
    "world" ,
    "large" ,
    "language" ,
    "models" ,
    "chatgpt"
]
word_to_id = {}
for i , word in enumerate(words) :
  word_to_id[word] = i
print(word_to_id)
```

**Create Embedding Layer**

```python
vocab_size = len(words)
embedding_dimension = 8
embedding = nn.Embedding(vocab_size , embedding_dimension)
print(embedding)
```

**Convert Word into ID**

```python
word = "language"
word_id = torch.tensor([word_to_id[word]])
print("Word : " , word)
print("ID : " , word_id.item())
```

**Get Embedding Vector**

```python
vector = embedding(word_id)
print("Embedding Vector:\n")
print(vector)
```

**Check Vector Shape**

```python
print("Shape : " , vector.shape)
```

**Convert Multiple Words**

```python
sentence = [
    "artificial" ,
    "intelligence" ,
    "is" ,
    "changing" ,
    "the" ,
    "world"
]
ids = []
for word in sentence :
  ids.append(word_to_id[word])
ids = torch.tensor(ids)
print(ids)
```

**Get Sentence Embeddings**

```python
sentence_vectors = embedding(ids)
print(sentence_vectors)
```

**Print Every Word With its Vector**

```python
for word in sentence :
  idx = torch.tensor([word_to_id[word]])
  vector = embedding(idx)
  print("=" * 40)
  print("Word : " , word)
  print("Vector : ")
  print(vector)
```

**Vocabulary Size & Embedding Size**

```python
print("Vocabulary Size : " , vocab_size)
print("Embedding Dimension : " , embedding_dimension)
print("Embedding Weight Shape : " , embedding.weight.shape)
```

**Display All Embeddings**

```python
print("All Learned Embeddings")
print(embedding.weight)
```
