**TEMPERATURE SAMPLING**

**Import Libraries**

```python
import torch
print("PyTorch Version : " , torch.__version__)
```

**Sample Vocabulary**

```python
words = [
    "python" ,
    "machine" ,
    "learning" ,
    "artificial" ,
    "intelligence"
]
logits = torch.tensor([5.0 , 3.0 , 2.0 , 1.0 , 0.5])
print("Words : " , words)
print("Logits : " , logits)
```

**Convert Logits to Probabilities**

```python
probabilities = torch.softmax(logits , dim = 0)
print("Probabilities:\n")
for word , prob in zip(words , probabilities) :
  print(f"{word:<15}:{prob:4f}")
```

**Greedy Decoding**

```python
predicted = torch.argmax(probabilities).item()
print("Greedy Prediction : ")
print(words[predicted])
```

**Temperature Function**

```python
def temperature_sampling(logits , temperature) :
  scaled_logits = logits / temperature
  probabilities = torch.softmax(scaled_logits , dim = 0)
  return probabilities
```

**Compare Different Temperatures**

```python
temperatures = [0.5 , 1.0 , 2.0]
for t in temperatures :
  probs = temperature_sampling(logits , t)
  print(f"\nTemperature = {t}")
  for word , prob in zip(words , probs) :
    print(f"{word:<15}:{prob:4f}")
```

**Generate Using Temperature**

```python
temperature = 1.5
probabilities = temperature_sampling(logits , temperature)
prediction = torch.multinomial(probabilities , 1).item()
print("Temperature : " , temperature)
print("Generated Word : " , words[prediction])
```

**Top-k-Sampling**

```python
k = 3
top_probs , top_indices = torch.topk(probabilities , k)
top_probs = top_probs / top_probs.sum()
selected = torch.multinomial(top_probs , 1).item()
predicted_word = words[top_indices[selected]]
print("Top-k = " , k)
print("Generated Word : " , predicted_word)
```

**Compare Generation Methods**

```python
print("Greedy Decoding : " , words[torch.argmax(probabilities).item()])
temperature = 2.0
probs = temperature_sampling(logits , temperature)
sample = torch.multinomial(probs , 1).item()
top_probs , top_indices = torch.topk(probs , 3)
top_probs = top_probs / top_probs.sum()
sample = torch.multinomial(top_probs , 1).item()
print("Top-k Sampling : " , words[top_indices[sample]])
```
