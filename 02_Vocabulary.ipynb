**VOCABULARY BUILDING**

**Input Text**

```python
text = """
Artificial Intelligence is changing the world.
Large Language Models understand human language.
ChatGPT is an example of an LLM.
"""
print("Original Text:\n")
print(text)
```

**Clean the Text**

```python
import string
text = text.lower()
clean_text = text.translate(str.maketrans(" " , " " , string.punctuation))
print(clean_text)
```

**Tokenize**

```python
tokens = clean_text.split()
print("Tokens:\n")
print(tokens)
```

**Build Vocabulary**

```python
vocabulary = sorted(set(tokens))
print("Vocabulary:\n")
print(vocabulary)
```

**Vocabulary Size**

```python
print("Vocabulary Size : " , len(vocabulary))
```

**Create Word -> ID Dictionary**

```python
word_to_id = {}
for index , word in enumerate(vocabulary) :
  word_to_id[word] = index
print(word_to_id)
```

**Create ID -> Word Dictionary**

```python
id_to_word = {}
for word , index in word_to_id.items() :
  id_to_word[index] = word
print(id_to_word)
```

**Convert Words into IDs**

```python
token_ids = []
for word in tokens :
  token_ids.append(word_to_id[word])
print("Word IDs:\n")
print(token_ids)
```

**Convert IDs Back to Words**

```python
decoded_words = []
for index in token_ids :
  decoded_words.append(id_to_word[index])
print(decoded_words)
```

**Unknown Token Example**

```python
unknown_word = "robot"
if unknown_word in word_to_id :
  print(word_to_id[unknown_word])
else :
  print(f'"{unknown_word}" is not in the vocabulary.')
```

**Display Vocabulary Table**

```python
print("Vocabulary Table\n")
for word , index in word_to_id.items() :
  print(f"{word:<15} ---> {index}")
```
