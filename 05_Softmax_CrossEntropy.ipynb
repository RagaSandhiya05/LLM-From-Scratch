**SOFTMAX & CROSS-ENTROPY LOSS**

**Import PyTorch**

```python
import torch
import torch.nn as nn
print("PyTorch Version : " , torch.__version__)
```

**Create Sample Logits**

```python
logits = torch.tensor([2.5 , 1.2 , 0.3])
print("Logits : ")
print(logits)
```

**Apply Softmax**

```python
probabilities = torch.softmax(logits , dim = 0)
print("Probabilities : ")
print(probabilities)
```

**Verify Probability Distribution**

```python
print("Sum of Probabilities : " , probabilities.sum().item())
```

**Predict the Most Likely Word**

```python
words = ["python" , "machine" , "learning"]
predicted_index = torch.argmax(probabilities).item()
print("Predicted Word : " , words[predicted_index])
```

**Cross-Entropy Loss**

```python
target = torch.tensor([1])
criterion = nn.CrossEntropyLoss()
loss = criterion(logits.unsqueeze(0) , target)
print("Cross-Entropy Loss : " , loss.item())
```

**Better Prediction**

```python
better_logits = torch.tensor([1.0 , 5.0 , 0.2])
loss = criterion(better_logits.unsqueeze(0) , target)
print("Loss for Better Prediction : " , loss.item())
```

**Compare Good vs Bad Predictions**

```python
bad_logits = torch.tensor([4.0 , 1.0 , 0.5])
good_logits = torch.tensor([1.0 , 4.5 , 0.3])
bad_loss = criterion(bad_logits.unsqueeze(0) , target)
good_loss = criterion(good_logits.unsqueeze(0) , target)
print("Bad Prediction Loss : " , bad_loss.item())
print("Good Prediction Loss : " , good_loss.item())
```

**Display All Probabilities**

```python
print("Probability Distribution:\n")
prob = torch.softmax(logits, dim=0)
for word, probability in zip(words, prob):
    print(f"{word:<10} : {probability.item():.4f}")
```

```python
sentence = "i love"
print("Input Sentence : " , sentence)
print("Predicted Next Word : " , words[predicted_index])
```
